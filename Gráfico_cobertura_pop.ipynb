Para começar a criação do gráfico eu faço a leitura da tabela apresentada acima, para fazer a leitura uso a bibliteca pandas que foi importada e declarada como pd:

```python
import pandas as pd
```

Faço a leitura dos dados e atribuo os mesmos a variável "dados":

```python
dados = pd.read_excel("https://github.com/gustavoalmeidam7/cobertura-pop-ambulancia/raw/refs/heads/main/cobertura_pop_ambulancia.xlsx")
```

Agora com os dados da tabela carregados na variável é possivel fazer a deserialização dos mesmos e converter de dataframe para os dados necessários:

```python
# Lambda para pegar o ano da linha i da tabela
filtarAnos = lambda i: int(dados.iloc[i].iloc[0])

# Lambda para pegar a cobertura da linha i da tabela
filtarCobertura = lambda i: round(float(dados.iloc[i].iloc[1]), 2)

# Executa função de mapeamento iterando sobre a quantidade de linhas
anos = list(map(filtarAnos, range(len(dados))))
coberturas_pop = list(map(filtarCobertura, range(len(dados))))

print(f"Anos: {anos} \nCobertura: {coberturas_pop}")
```

Cos os dados devidamente formatados agora é possível começar a plotar o gráfico utilizando a biblioteca matplotlib, abaixo eu a importo e atribuo o nome como "plt":

```python
import matplotlib.pyplot as plt
```

Com a biblioteca devidamente importada agora pode ser utilizada, foi definido um gráfico de linhas e alimentado pelos dados da planilha:

```python
# Define o tamanho da figura
plt.figure(figsize=(7,7))
# Plota as linhas no grafico
plt.plot(anos, coberturas_pop, marker='o', color='green')

# Adiciona textos de porcentagem nos respectivos pontos
for i, val in enumerate(coberturas_pop):
  plt.text(anos[i], val + 0.8, str(f"{val}%"), ha='center', va='bottom')

plt.title('Percentual de cobertura populacional dos serviços de ambulância')

plt.xlabel('Ano')
plt.ylabel('Percentual (%)')

plt.grid(True)
plt.show()

```
